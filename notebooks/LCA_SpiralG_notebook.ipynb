## 2) Import the databases & perform LCA calculations

**Database import:** The first step consists of selecting or creating a Brightway project. The project is located in the working directory selected above. The two background databases *biosphere3* and *ecoinvent* (here the version 3.6 cut-off) are then imported. In this study, the foreground databases used are imported as BW2package files (see folder "databases"). These databases were generated using the SpiralG biorefinery model (see folder "bioref_model") for periods 1 & 2. Each period corresponds to an independent data collection campaign conducted at the three pilot scale facilities located in Italy (S1) and France (S2 & S3). Two independent (and non-parameterised) sets of databases were created to compare the environmental impacts for the two periods. 

**Contribution analysis:** Three types of contribution analysis can be performed to compare the subsystems (e.g., S1, S2, and S3), exchange categories (e.g., nutrients, energy, transport), and activities within a subsystem (e.g. S1.A1.Cultivation, S1.A2.Filtration). The type of analysis to perform can be selected in the recipe by manually settig a boolean (e.g., True or False). To demonstrate the different functionalities, the analysis are performed one by on below (i.e. each calculation step corresponds to a specific recipe).
