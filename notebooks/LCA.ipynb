# Import the databases & perform LCA calculations

**Database import:** The first step consists of selecting or creating a Brightway project. The project is located in the working directory selected above. The two background databases *biosphere3* and *ecoinvent* (here the version 3.6 cut-off) are then imported. In this study, the foreground databases used are imported as BW2package files (see folder "databases"). These databases were generated using the SpiralG biorefinery model (see folder "bioref_model") for periods 1 & 2. Each period corresponds to an independent data collection campaign conducted at the three pilot scale facilities located in Italy (S1) and France (S2 & S3). Two independent (and non-parameterised) sets of databases were created to compare the environmental impacts for the two periods. 

**Contribution analysis:** Three types of contribution analysis can be performed to compare the subsystems (e.g., S1, S2, and S3), exchange categories (e.g., nutrients, energy, transport), and activities within a subsystem (e.g., S1.A1.Cultivation, S1.A2.Filtration). The type of analysis to perform can be selected in the recipe by manually setting a boolean (e.g., True or False). To demonstrate the different functionalities, the analysis are performed one by one below (i.e. each calculation step corresponds to a specific recipe).

# Local sensitivity analysis: FU & allocation procedure

## Sensitivity of the LCA results to the functional unit

In the main analysis, a fixed composition of the set of biorefinery products was used to compare periods 1 and 2 from a technological point of view. Here, the LCA calculation are performed using the FU "28.78 kg DW-eq of Spirulina dry spaghettini" (FU1-2) and the outcomes are compared with the results obtained for FU1-1 "Production of 28.78 kg DW-eq of Spirulina dry spaghettini and processing into 5.70 kg DW-eq of blue extract, 3.46 kg DW-eq of CPAc, and 8.51 kg DW-eq of CPBc."

The calculations were performed using the inventory data for period 1. The results obatined in the main analysis (i.e. using the data for period 1, FU1-1, and the allocation factors for period 2) are compared to the results obtained using period 1, FU1-2, and the allocation factors for period 1. 

This approach allows to evaluate the influence of the composition of the set of boprefinery products on the LCA results. This is particularly relevant to consider when seasonal variations influence the biochemical composition of the biomass. 

The datasets for S2 and S3 need to be rescaled to 4.86 (instead of 5.70) and 1.36 instead of 3.47 kg DW-eq, respectively. The rescaling is accounted for in the recipe file. The calculations are performed for S123 using FU1-1 and FU2-1. No allocation is applied.

```python
from master import prep_dir, import_databases, run

recipe = prep_dir (wdir = wdir, rdir = rdir, recipe_name = 'recipe_sensitivity_FU.yml')
recipe = import_databases (recipe, wdir)
recipe = run (recipe, wdir, dsdir)
```
